```python
import matplotlib.pyplot as plt
import numpy as np

# Set the style to 'ggplot' to match the grey background and gridlines
plt.style.use('ggplot')

save_path = "/Users/apple/mlpractical/figures"
file_name = 'label-smoothing'

# Sample data from the table (as per the given structure)
dropout_values = [0.6, 0.7, 0.85, 0.97]
dropout_val_acc = [0.807, 0.830, 0.851, 0.854]
dropout_gap = [0.044, 0.158, 0.105, 0.213]

weight_decay_values = [5e-4, 1e-3, 5e-3, 5e-2]
L1_val_acc = [0.795, 0.764, 0.241, 0.220]
L1_gap = [0.016, -0.018, 0.0, 0.0]
L2_val_acc = [0.851, 0.855, 0.813, 0.392]
L2_gap = [0.154, 0.064, 0.021, -0.002]

# Create the first figure for dropout values
fig_1 = plt.figure(figsize=(8, 4))
ax_1 = fig_1.add_subplot(111)
ax_1.plot(dropout_values, dropout_val_acc, label="Val. Acc.")
ax_1.set_xlabel('Dropout value')
ax_1.set_ylabel('Accuracy')
ax_1.set_ylim(0.8, 0.86)  # Adjust the y-axis for accuracy to focus on relevant range
ax_1.set_xlim(0.5, 1.0)

ax_2 = ax_1.twinx()
ax_2.plot(dropout_values, dropout_gap, label="Gap", color ='C1')
ax_2.set_ylabel('Generalization gap')
ax_2.set_ylim(0, 0.25)  # Adjust the y-axis for the generalization gap to the relevant range

ax_1.legend(loc="upper left")
ax_2.legend(loc="upper right")
fig_1.savefig(save_path + f'/{file_name}_droupout.pdf')


# Create the second figure for weight decay values
fig_2 = plt.figure(figsize=(8, 4))
ax_12 = fig_2.add_subplot(111)
ax_12.plot(weight_decay_values, L1_val_acc, label="L1 Val. Acc.")
ax_12.plot(weight_decay_values, L2_val_acc, label="L2 Val. Acc.")
ax_12.set_xlabel('Weight decay value')
ax_12.set_ylabel('Accuracy')
ax_12.set_xscale('log')
ax_12.set_ylim(0.2, 0.9)  # Adjust the y-axis for accuracy to fit the range of values

ax_22 = ax_12.twinx()
ax_22.plot(weight_decay_values, L1_gap, label="L1 Gap", linestyle='--')
ax_22.plot(weight_decay_values, L2_gap, label="L2 Gap", linestyle='--')
ax_22.set_ylabel('Generalization gap')
ax_22.set_ylim(-0.05, 0.2)  # Adjust the y-axis for the generalization gap to the relevant range

ax_12.legend(loc="upper left")
ax_22.legend(loc="upper right")
fig_2.savefig(save_path + f'/{file_name}_weight-decay.pdf')


plt.tight_layout()
plt.show()
```
